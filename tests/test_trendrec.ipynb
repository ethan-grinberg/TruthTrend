```python
import pandas as pd
from pytrends.request import TrendReq
test_keyword = "Blockchain"
```

```python
pytrends = TrendReq(hl='en-US', tz=360)
pytrends.build_payload([test_keyword], cat=0, timeframe='today 5-y', geo='', gprop='')
```

```python
pytrends.related_topics()[test_keyword]["rising"]
```

```python
pytrends.related_topics()[test_keyword]["top"]
```

```python
pytrends.related_queries()[test_keyword]["top"]
```

```python
pytrends.suggestions(test_keyword)
```

```python
pytrends.trending_searches()
```

```python
pytrends.categories()
```
